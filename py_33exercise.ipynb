```python
#ex 19
# Get the input strings
original_string = input()
substring = input()

# Count the occurrences of the substring in the original string
count = original_string.count(substring)

# Initialize a counter variable
count = 0

# Iterate through the original string
for i in range(len(original_string) - len(substring) + 1):
    # Check if the substring matches the current slice of the original string
    if original_string[i:i+len(substring)] == substring:
        # Increment the counter if there is a match
        count += 1

# Print the result
print(count)
```

```python
#20 
from collections import OrderedDict

ordinary_dictionary = OrderedDict()

for _ in range(int(input())):

    data = input().split()
    price = int(data.pop())
    name = ' '.join(data)

    if name in ordinary_dictionary:
        ordinary_dictionary[name] += price
    else:
        ordinary_dictionary[name] = price

[print(item[0] + ' ' + str(item[1])) for item in ordinary_dictionary.items()]
    
```

```python
#21
#21

S = input()


contains_alnum = any(char.isalnum() for char in S)


contains_alpha = any(char.isalpha() for char in S)


contains_digit = any(char.isdigit() for char in S)


contains_lower = any(char.islower() for char in S)

contains_upper = any(char.isupper() for char in S)

print(contains_alnum) 
print(contains_alpha)
print(contains_digit) 
print(contains_lower) 
print(contains_upper)
```

```python
#22
#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must me an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))  
```
